```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Set random seed for reproducibility
np.random.seed(0)

# Number of mock patients
n = 100

# Create DataFrame with mock values
df = pd.DataFrame({
    "patient_id": range(1, n + 1),
    "systolic": np.random.normal(120, 15, n).round(),
    "diastolic": np.random.normal(80, 10, n).round()
})

print(df.head())

# Plot histograms
plt.figure(figsize=(12, 5))

# Systolic histogram
plt.subplot(1, 2, 1)
plt.hist(df["systolic"], bins=15, color="steelblue", edgecolor="black")
plt.title("Systolic Blood Pressure Distribution")
plt.xlabel("Systolic (mmHg)")
plt.ylabel("Frequency")

# Diastolic histogram
plt.subplot(1, 2, 2)
plt.hist(df["diastolic"], bins=15, color="salmon", edgecolor="black")
plt.title("Diastolic Blood Pressure Distribution")
plt.xlabel("Diastolic (mmHg)")
plt.ylabel("Frequency")

plt.tight_layout()
plt.show()

```
